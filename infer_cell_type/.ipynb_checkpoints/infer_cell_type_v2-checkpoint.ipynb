```python
f = open('TS_X.pickle', 'rb')
matrix = pickle.load(f)
f.close()
```

```python
from sklearn.decomposition import PCA
principal=PCA(n_components=2000)
```

```python
%%time
pca_mat = principal.fit(matrix)
```

```python
with open('TS_pca.pickle', 'wb') as f:
    pickle.dump(pca_mat, f)
    f.close()
```

```python
eigenVec = pca_mat.components_
eigenVal = pca_mat.explained_variance_
```

```python
eigenVec
```

```python
print(principal.explained_variance_ratio_)
```
